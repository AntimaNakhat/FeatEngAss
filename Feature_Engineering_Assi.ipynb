1. What is a parameter?

A parameter is a configuration variable within a machine learning model that is learned from the data during training. For example, weights and biases in neural networks are parameters. These are adjusted during the training process using optimization techniques to minimize the loss function and improve the model's accuracy.

2. What is correlation?

What does negative correlation mean?


Correlation measures the statistical relationship between two variables. It indicates the degree to which the variables move in relation to each other. The value of correlation ranges from -1 to 1, where 1 signifies a perfect positive relationship, -1 signifies a perfect negative relationship, and 0 indicates no relationship.


Negative correlation occurs when an increase in one variable leads to a decrease in the other. For example, if study time increases, errors in an exam might decrease, showing a negative correlation.

3. Define Machine Learning. What are the main components in Machine Learning?

Machine Learning is the process of enabling systems to learn patterns from data without explicit programming. Key components include:

Data: Raw information to train the model.

Features: Variables that represent the data.

Model: An algorithm to learn from data.

Loss Function: Measures model error.

Optimizer: Updates parameters to minimize error.

4. How does loss value help in determining whether the model is good or not?

The loss value quantifies the error between predicted and actual outcomes. A lower loss value indicates better model performance. If the loss value is high, it signals underfitting, meaning the model needs improvement in learning from data.

5. What are continuous and categorical variables?

Continuous variables can take any numerical value within a range (e.g., height, weight). Categorical variables represent distinct categories (e.g., gender, colors).

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

Categorical variables can be encoded using:

One-Hot Encoding: Converts categories into binary vectors.
Label Encoding: Assigns integers to categories.
Ordinal Encoding: Preserves category order if applicable.

7. What do you mean by training and testing a dataset?

Training a dataset involves teaching a model on labeled data, while testing evaluates its performance on unseen data. This ensures the model generalizes well.

8. What is sklearn.preprocessing?

sklearn.preprocessing provides tools to preprocess data, such as scaling, normalization, encoding categorical variables, and handling missing values.

9. What is a Test set?

A test set is a subset of data not used in training, reserved for evaluating model performance.

15. What is causation? Explain the difference between correlation and causation with an example.

Causation implies one variable directly affects another, while correlation only shows a relationship. For instance, ice cream sales and drowning incidents correlate but aren't causally linked; temperature affects both.

16. What is an Optimizer? What are different types of optimizers?

An optimizer updates model parameters to minimize loss. Examples include:

Gradient Descent: Basic optimizer adjusting weights by loss gradient.
Adam: Combines momentum and adaptive learning rates.
